## 2.0 Read Customed Lexicon Files
Read the lexicon sentiment files created from Assignment 1.
Download it from [lexicon_political_master.csv](https://drive.google.com/drive/folders/1iMEVnEUxr-JtUaxuZfHxd3Shnch6149A?usp=sharing)

```
with open('lexicon_political_master.csv', 'r') as f:
    reader = csv.reader(f)
    lexicon = list(reader)
```

Functions that can be reused

* assess_review_file - Function to process the Reviews file
* assess_accuracy - Function to check the accurary



```
# READ THE REVIEW FILES AND COUNT THE SENTIMENT 
# RETURN COLUMN <Positive/Negative Source File, Text, Positive Count, Negative Count, Total Count>
def assess_review_file(file_name, file_type):
  items = []
  with open(file_name, 'r') as reviewfile:
    reader = csv.reader(reviewfile)
    next(reader, None)
    item = []

    #print("----- VALIDATE INDIVIDUAL REVIEW -----")

    for row in reader:
      positive_ttl = 0
      negative_ttl = 0
      ttl = 0
      review = row[0].lower()
      
      print("Review: {}".format(review))
      word_tokens = nltk.word_tokenize(review.lower())
      for word in word_tokens:
        res = [i for i in lexicon if word in i] 
        if len(res) > 0:
          score = int(res[0][1])
          ttl += score
          print("{}: {}".format(word, score))
          if score > 0:
            positive_ttl += 1
          elif score < 0:
            negative_ttl += -1

      print("Review Positive Score: {}".format(positive_ttl))
      print("Review Negative Score: {}".format(negative_ttl))
      print("Review Total Score: {}\n".format(ttl))

      item = [file_type, review, positive_ttl, negative_ttl, (positive_ttl + negative_ttl)]
      items.append(item)
    
    return items

def assess_accuracy(reviews):
  print("\n\n------------ DOCUMENT RESULT ------------")
  total_review = len(reviews)
  total_positive = 0
  total_negative = 0
  total_neutral = 0
  for ele in enumerate(reviews): 
    if ele[1][4] > 0:
      total_positive += 1 
    elif ele[1][4] < 0:
      total_negative += 1
    elif ele[1][4] == 0:
      total_neutral += 1

  print("Total Reviews: {}".format(total_review))
  print("Total Reviews Flagged as Neutral: {}".format(total_neutral))
  print("Total Reviews Flagged as Positive: {}".format(total_positive))
  print("Total Reviews Flagged as Negative: {}".format(total_negative))
  print("Posaitive Accuracy Rate: {}%".format(round((total_positive/total_review)*100, 2)))
  print("Negative Accuracy Rate: {}%".format(round((total_negative/total_review)*100, 2)))
```

## 3.0 Process Review files and Review the Accuracy
### 3.1 Positive File

```
PositiveReviews = assess_review_file("politic_issues_positive_reviews.csv", "PositiveFile")
assess_accuracy(PositiveReviews)
```

### 3.2 Negative File

```
NegativeReviews = assess_review_file("politic_issues_negative_reviews.csv", "NegativeFile")
assess_accuracy(NegativeReviews)
```
